```python
# cleaning NaN entries in the train data
def removeNanQuant(num):
    if(math.isnan(num)):
        return 0.0
    return num

def removeNanItem(item):
    if(type(item) is not str and math.isnan(item)):
        return "None"
    return item

cols = list(train.columns)

for col in cols:
    if(col[0:9] == "quantity_"):
        train[col] = train[col].apply(removeNanQuant)
    elif(col[0:5] == "item_"):
        train[col] = train[col].apply(removeNanItem)

# checks for number of nan values equal to zero.
for col in train.columns:
    if(col[0:9] == "quantity_"):
        print(train[col].isna().sum())
    elif(col[0:5] == "item_"):
        print(train[col].isna().sum())

train.head()
```

```python
# cleaning NaN entries in the test data

cols = list(test.columns)

for col in cols:
    if(col[0:9] == "quantity_"):
        test[col] = test[col].apply(removeNanQuant)
    elif(col[0:5] == "item_"):
        test[col] = test[col].apply(removeNanItem)

# checks for number of nan values equal to zero.
for col in test.columns:
    if(col[0:9] == "quantity_"):
        print(test[col].isna().sum())
    elif(col[0:5] == "item_"):
        print(test[col].isna().sum())
test
```

```python
from dateutil.parser import parse

def get_datetime(string):
    return parse(string)
time_objs = list(map(get_datetime, train["datetime"]))

def get_year(obj):
    return obj.year

def get_month(obj):
    return obj.month

def get_day(obj):
    return obj.day

def get_hour(obj):
    return obj.hour

def get_minute(obj):
    return obj.minute

def get_second(obj):
    return obj.second

year = list(map(get_year, time_objs))
months = list(map(get_month, time_objs))
days = list(map(get_day, time_objs))
hours = list(map(get_hour, time_objs))
minutes = list(map(get_minute, time_objs))
seconds = list(map(get_second, time_objs))

```

```python
# check if all the years are 2019

def checkEqual(list):
    return all(elem == list[0] for elem in list)

is_2019 = checkEqual(year)
is_2019
```

```python
# check if all the months are September

is_sept = checkEqual(months)
is_sept
```

```python
# check if all the days are the first seven days of September

def check_week(list):
    return all(elem == 1 or elem == 2 or elem == 3 or elem == 4 or elem == 5 or elem == 6 or elem == 7 for elem in list)

is_first_week = check_week(days)
is_first_week
```

```python
train["day"] = days
train["hour"] = hours
train["minute"] = minutes
train["second"] = seconds
```
